Atividade 2 - Consultar atendimentos com valor entre 90 e 200 reais.

```python
atendimentos_entre_90_200 = list(colecao.find({
    'valor': {'$gte': 90, '$lte': 200}
}))

display(atendimentos_entre_90_200)
```

Atividade 3 -  Consultar atendimentos das cidades Assis, Ourinhos, Pompeia e Marília.

```python
atendimentos_cidades = list(colecao.find({
    'cidade': {'$in': ['Assis', 'Ourinhos', 'Pompeia', 'Marilia']}
}))

display(atendimentos_cidades)
```

Atividade 4 - Converter os resultados para DataFrame.

```python
df_atendimentos_90_200 = pd.DataFrame(atendimentos_entre_90_200)
display(df_atendimentos_90_200)
```

```python
df_cidades = pd.DataFrame(atendimentos_cidades)
display(df_cidades)
```

Atividade 5 - Verificar valores nulos.

```python
df_atendimentos_90_200.isnull().sum()
```

```python
df_cidades.isnull().sum()
```

Atividade 6 - Calcular faturamento total.

```python
faturamento_total_atendimentos_90_200 = df_atendimentos_90_200['valor'].sum()
print('Faturamento total: ', faturamento_total_atendimentos_90_200)
```

```python
faturamento_total_cidades = df_cidades['valor'].sum()
print('Faturamento total: ', faturamento_total_cidades)
```

Atividade 7 -  Calcular faturamento por cidade.

```python
faturamento_por_cidade = df_cidades.groupby('cidade')['valor'].sum().sort_values(ascending=False)
display(faturamento_por_cidade)
```

Atividade 8 - Calcular faturamento por serviço.

```python
faturamento_por_servico = df_cidades.groupby('servico')['valor'].sum().sort_values(ascending=False)
display(faturamento_por_servico)
```

Atividade 9 - Calcular ticket médio geral

```python
ticket_medio_geral = df_cidades['valor'].mean()
print('Ticket médio geral: ', ticket_medio_geral)
```

Atividade 10 -  Calcular ticket médio por cidade


```python
ticket_medio_por_cidade = df_cidades.groupby('cidade')['valor'].mean().round(2)
print(ticket_medio_por_cidade)
```

Atividade 11 - Identificar o serviço mais lucrativo.

```python
servico_top = df_cidades.groupby('servico')['valor'].sum().idxmax()
print('O serviço mais lucrativo é: ', servico_top) 
```

Atividade 12 - Identificar a cidade com maior faturamento

```python
cidade_top = df_cidades.groupby('cidade')['valor'].sum().idxmax()
print('A cidade mais lucrativa é: ', cidade_top)
```

Atividade 13 -  Identificar os 3 maiores atendimentos

```python
top_3_atendimentos = df_cidades.groupby('servico')['valor'].sum().sort_values(ascending=False).head(3)
display(top_3_atendimentos)
```

Relatório - O relatório deve ser gerado como um dicionário Python e conter EXATAMENTE os seguintes campos:

· data_geracao (data/hora atual)
· faturamento_total (float)
· faturamento_por_cidade (dict)
· faturamento_por_servico (dict)
· ticket_medio_geral (float)
· ticket_medio_por_cidade (dict)
· servico_mais_lucrativo (string)
· cidade_maior_faturamento (string)
· top_3_atendimentos (lista de registros com maior valor)

```python
relatorio = {
    'data_geracao': datetime.now(),
    'faturamento_total': float(faturamento_total_cidades),
    'faturamento_por_cidade': faturamento_por_cidade.to_dict(),
    'faturamento_por_servico': faturamento_por_servico.to_dict(),
    'ticket_medio_geral': float(ticket_medio_geral),
    'ticket_medio_por_cidade': ticket_medio_por_cidade.to_dict(),
    'servico_mais_lucrativo': servico_top,
    'cidade_maior_faturamento': cidade_top,
    'top_3_atendimentos': top_3_atendimentos
}

display(relatorio)
```
